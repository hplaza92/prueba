## Columnas de características de preproceso
Como puede verse, hay varias columnas no numéricas que deben convertirse. Muchas de ellas son simplemente columnas del tipo: sí/no, por ejemplo: Internet. Estas variables se pueden convertir razonablemente en valores 1/0 (binarios).

Otras columnas, como Mjob y Fjob, tienen más de dos valores y se conocen como variables categóricas. La forma recomendada de manejar una columna de este tipo es crear tantas columnas como sea posible los valores (por ejemplo, Fjob_teacher, Fjob_other, Fjob_services, etc.) y asignar un 1 a una de ellas y 0 a todas las demás.

Estas columnas generadas a veces se denominan variables "dummy", y utilizaremos la función [pandas.get_dummies()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html) para realizar esta transformación. Ejecute la celda de código a continuación para realizar la rutina de preprocesamiento que se describe en esta sección.


````
def preprocess_features(X):
    ''' Preprocesa el dataset y convierte las variables binarias a (0/1). Convierte las variables categóricas en variables dummy. '''

    # Crea un nuevo dataframe con el index del dataframe que uses de entrada
    output = pd.DataFrame(index = X.index)

    # Investiga cada columna del dataset de entrada
    for col, col_data in X.iteritems():

        # Si el tipo de dato es non-numeric, reemplaza el valor yes/no por 1/0
        if col_data.dtype == object:
            col_data = col_data.replace(['yes', 'no'], [1, 0])

        # Si el dato es categorico, conviertelo a variables dummy
        if col_data.dtype == object:
            # Ejemplo: 'school' => 'school_GP' y 'school_MS'
            col_data = pd.get_dummies(col_data, prefix = col)

        # Añade la columna "procesada" al dataset de salida
        output = output.join(col_data)

    return output

X_all = preprocess_features(X_all)
print("Lista de variables de entrenamiento preprocesado ({} en total):\n{}".format(len(X_all.columns), list(X_all.columns)))
````

Esta función informará el puntaje F1 para los datos de entrenamiento y prueba por separado.

````
def train_classifier(clf, X_train, y_train):
    ''' Entrena el algoritmo. '''

    # Comienza a medir el tiempo, entrena en algoritmo y para el reloj
    start = time()
    clf.fit(X_train, y_train)
    end = time()

    # Imprime los resultados
    print("Entrenamiento en {:.4f} segundos".format(end - start))


def predict_labels(clf, features, target):
    ''' Genera predicciones y mide la calidad usando el f1 score. '''

    # Comienza a medir el tiempo, genera las predicciones y para el reloj
    start = time()
    y_pred = clf.predict(features)
    end = time()

    # Imprime los resultados y devuelve las metricas
    print("Predicciones generadas en {:.4f} segundos.".format(end - start))
    return f1_score(target.values, y_pred, pos_label='yes')


def train_predict(clf, X_train, y_train, X_test, y_test):
    ''' Entrena y predicde en función de F1 score. '''

    # Imprime el nombre del algoritmo y el tamaño del training set
    print("")
    print("Entrenando un {} usando un dataset de entrenamiento de {}. . .".format(clf.__class__.__name__, len(X_train)))

    # Entrena el algoritmo
    train_classifier(clf, X_train, y_train)

    # Imprime los resultados sobre train y test
    print("F1 score sobre el conjunto de entrenamiento: {:.4f}.".format(predict_labels(clf, X_train, y_train)))
    print("F1 score sobre el conjunto de validación: {:.4f}.".format(predict_labels(clf, X_test, y_test)))
````
